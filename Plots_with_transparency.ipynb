# This notebook gives an example of how to create plots with transparency
## And save it as a png file

**Loading necessary packages. Please install packages if not present.**

```python
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.cm as cm
```

**Creating data to plot**

```python
x=np.linspace(-10,10,num=100)

amp=5
d=2

y=amp*np.sin(x)
yu=(amp+d)*np.sin(x)
yd=(amp-d)*np.sin(x)
```

**Making the plot and saving**

```python
plt.figure()
plt.plot(x,y)
plt.fill_between(x, yu, yd, color='cyan', alpha=.2)
plt.savefig('sinewtransparency.png',dpi=300,format='png')
plt.show()
```
